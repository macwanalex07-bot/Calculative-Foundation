```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

# Dataset
data = np.array([
    [80, 70, 75],
    [60, 65, 70],
    [90, 85, 88]
])

# Convert to DataFrame
df = pd.DataFrame(data, columns=["Math", "Physics", "Chemistry"])

# Norms
norms = np.linalg.norm(data, axis=1)

# Dot product
dot_product = np.dot(data[0], data[1])

# Cross product
cross_product = np.cross(data[0], data[1])

# Projection
projection = (dot_product / np.dot(data[1], data[1])) * data[1]

# Matrix operations
transpose = data.T
determinant = np.linalg.det(data)
inverse = np.linalg.inv(data)

# Eigenvalues
cov = np.cov(data.T)
eigenvalues, eigenvectors = np.linalg.eig(cov)

# PCA
pca = PCA(n_components=2)
pca_result = pca.fit_transform(data)

# LDA (dummy labels)
labels = [1, 0, 1]  # Above / Below avg
lda = LinearDiscriminantAnalysis(n_components=1)
lda_result = lda.fit_transform(data, labels)

print("Norms:", norms)
print("Dot Product:", dot_product)
print("Cross Product:", cross_product)
print("Projection:", projection)
print("Determinant:", determinant)
print("Eigenvalues:", eigenvalues)
print("PCA Result:\n", pca_result)
print("LDA Result:\n", lda_result)
```
